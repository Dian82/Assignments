```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([421,1128,2596,8402,14133,23094,90158,147631])
p=np.array([4400,4700,7000,9800,16100,23300,247700,121800])/1000
s=np.array([2900,5800,21500,30600,52800,76200,255300,457300])/1000
plt.plot(x,s,"x",color="red",label="Sequential")
plt.plot(x,s,color="coral")
plt.plot(x,p,"^",color="#41b6c4",label="Parallel")
plt.plot(x,p,color="#41b6c4")
plt.xlabel("# of non-zeros")
plt.ylabel("Execution time ($\mu$s)")
plt.legend()
plt.savefig(r"size.pdf",format="pdf")
plt.show()
```
