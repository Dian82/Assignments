```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
res = np.load("results/simple-ai-lenet5.npz",allow_pickle=True)
```

```python
plt.plot(res["train_acc"],label="Train")
plt.plot(res["eval_acc"],label="Test")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
plt.plot(res["train_loss"],label="Train")
plt.plot(res["eval_loss"],label="Test")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()
```
