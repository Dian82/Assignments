```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_acc(res,network,max_iter=100):
    plt.plot(res["train_acc"][:max_iter+1],label="Train")
    plt.plot(res["eval_acc"][:max_iter+1],label="Test")
    if max_iter == 100:
        plt.axvline(x=13, color='r', linestyle='--')
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.title(network)
    plt.legend()
    plt.savefig("../report/fig/{}_acc.pdf".format(network))
    plt.show()
```

```python
def plot_loss(res,network,max_iter=100):
    plt.plot(res["train_loss"][:max_iter+1],label="Train")
    plt.plot(res["eval_loss"][:max_iter+1],label="Test")
    if max_iter == 100:
        plt.axvline(x=13, color='r', linestyle='--')
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title(network)
    plt.legend()
    plt.savefig("../report/fig/{}_loss.pdf".format(network))
    plt.show()
```

```python
res_lenet = np.load("results/simple-ai-LeNet5.npz",allow_pickle=True)
res_vgg = np.load("results/simple-ai-VGG16.npz",allow_pickle=True)
plot_acc(res_lenet,"LeNet5")
plot_loss(res_lenet,"LeNet5")
plot_acc(res_vgg,"VGG16",15)
plot_loss(res_vgg,"VGG16",15)
```

```python
plt.plot(res_lenet["eval_acc"][:16],label="LeNet5")
plt.plot(res_vgg["eval_acc"][:16],label="VGG16")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.savefig("../report/fig/acc_comparison.pdf")
plt.show()
```
